```python
from PIL import Image, ImageDraw, ImageFont
import csv
import os

# Load the CSV file
csv_file_path = "session.csv"

# Load the original template image
original_template_path = "img/secert.jpeg"
original_template = Image.open(original_template_path)

# Set the font for full names
name_font = ImageFont.truetype("Butter-Cookies.ttf", 120)  # You can change this to your desired font

# Ensure the output directory exists
output_directory = "img/cert"
os.makedirs(output_directory, exist_ok=True)

# Read data from CSV and create certificates with capitalized full names
with open(csv_file_path, 'r', newline='', encoding='utf-8', errors='ignore') as csvfile:
    reader = csv.DictReader(csvfile)
    for row in reader:
        full_name = row["userName"].upper()  # Capitalize each word in the full name

        # Create a copy of the original template image
        template_image = original_template.copy()

        # Create a drawing object for the current template image
        draw = ImageDraw.Draw(template_image)

        # Calculate the position to center the name text above the image
        text_bbox = draw.textbbox((0, 0), full_name, font=name_font)
        text_width = text_bbox[2] - text_bbox[0]
        center_x = (template_image.width - text_width) // 2
        name_text_position = (center_x, 540)  # Adjusted position above the image

        # Draw the name text
        name_text_color = (0, 0, 0)  # Set your desired text color
        draw.text(name_text_position, full_name, fill=name_text_color, font=name_font)

        # Save each certificate with a unique name
        output_image_path = os.path.join(output_directory, f"{full_name}.pdf")
        template_image.save(output_image_path, format="PDF")
```

```python
import csv
import os
import fitz  # PyMuPDF

# Load the CSV file
csv_file_path = "session.csv"

# Load the original template PDF
original_template_path = "img/cert.pdf"

# Ensure the output directory exists
output_directory = "img/cert"
os.makedirs(output_directory, exist_ok=True)

# Read data from CSV and create certificates with full names
with open(csv_file_path, 'r', newline='', encoding='utf-8', errors='ignore') as csvfile:
    reader = csv.DictReader(csvfile)
    for row in reader:
        full_name = row["userName"]  # Assuming full name is in the "userName" column of CSV, adjust if needed

        # Load the PDF template
        pdf_document = fitz.open(original_template_path)
        
        # Iterate through each page in the PDF template
        for page_num in range(pdf_document.page_count):
            page = pdf_document[page_num]

            # Create a copy of the page
            page_copy = page.copy()

            # Create a drawing object for the current page
            draw = page_copy.new_draw()

            # Calculate the position to center the name text above the image
            text_bbox = draw.textbbox((0, 0), full_name)
            text_width = text_bbox[2] - text_bbox[0]
            center_x = (page_copy.rect.width - text_width) // 2
            name_text_position = (center_x, 440)  # Adjusted position above the image

            # Draw the name text
            name_text_color = (0, 0, 0)  # Set your desired text color
            draw.text(name_text_position, full_name, color=name_text_color)

            # Save each certificate with a unique name
            output_image_path = os.path.join(output_directory, f"{full_name}_page_{page_num + 1}.pdf")
            page_copy.save(output_image_path)

        # Close the PDF document
        pdf_document.close()
```
